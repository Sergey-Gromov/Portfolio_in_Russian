# Прогнозирование оттока клиентов телеком компании

**Цель проекта:**  

Разработка модели модели машинного обучения для оператора связи «Ниединогоразрыва.ком», позволяющей прогнозировать отток клиентов.  
Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия.  
Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

Эффективность работы модели будет оцениваться метрикой ROC-AUC.  
Модель должна показать на тестовой выборке уровень метрики ROC-AUC не менее 0.85

**Описание услуг оператора**  

Оператор предоставляет два основных типа услуг:
1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. digital subscriber line, «цифровая абонентская линия») или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

**Описание предоставленных данных**

Данные состоят из файлов, полученных из разных источников:
- contract_new.csv — информация о договоре
- personal_new.csv — персональные данные клиента
- internet_new.csv — информация об интернет-услугах
- phone_new.csv — информация об услугах телефонии

Во всех файлах столбец customerID содержит код клиента.  
Информация о договорах актуальна на 1 февраля 2020.

## Предобработка данных

### Заполнение пропусков и коррекция типов данных

Заполняем пропуски флагом 'Not_using', что означает, что пользователь не использовал эти услуги

```python
df_merged = df_merged.fillna(value='Not_using')

# контроль результата
print(df_merged.shape)
print(df_merged.info())
display(df_merged.head())
```

Признак в колонке 'SeniorCitizen' по форме числовой, но по факту категориальный, даже бинарный. Поэтому его лучше привести к статндартному типу Yes/No

```python
df_merged.loc[df_merged['SeniorCitizen'] == 1, 'SeniorCitizen'] = 'Yes'
df_merged.loc[df_merged['SeniorCitizen'] == 0, 'SeniorCitizen'] = 'No'

# контроль результата
print(sorted(df_merged['SeniorCitizen'].unique())) 
print(df_merged.info())
display(df_merged.head())
```

### Проверка признаков на мультиколлинеарность

Определим функцию для вывода матрицы корреляции, построенной методом phik

```python
def corr_matrix_phik(data):
    
    # построение матрицы корреляции
    corr_matrix = phik_matrix(
                        data,
                        verbose=False,
                        njobs=-1,
    )

    # маска для вывода нижней половины матрицы
    mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
    np.fill_diagonal(mask, False)   

    # вывод матрицы в виде Heatmap
    plt.figure(figsize=(12,10), dpi= 80)
    sns.heatmap(corr_matrix,
                mask=mask, 
                xticklabels=corr_matrix.columns, 
                yticklabels=corr_matrix.columns, 
                cmap='RdYlGn', 
                center=0, 
                annot=True
    )
    plt.title('Корреляция признаков', fontsize=22)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.show()
```

Построим матрицу корреляции для изначального набора признаков

```python
# список категориальных признаков для последующего кодирования.
cat_features = features_train.select_dtypes(include='object').columns.to_list()
print(cat_features)
```

## Постройка и обучение  нескольких моделей

Будем оценивать качество обучения на тренировочной выборке с применением кросс-валидации и подбором гиперпараметров для трёх моделей, построенных на разных принципах:

- модели Логистической регрессии - это линейная модель классификации  
- модели RandomForestClassifier - это модель классификации на основе деревьев  
- модели CatBoostClassifier - это модель классификации на основе деревьев с применением градиентного бустинга

Для реализации подбора гиперпараметров на кросс-валидации будем использовать конвейер scikit-learn pipeline. Это предотвратит утечку при кросс-валидации.

В результате будет выбрана лучшая модель, которая потом будет проверяться на тестовой выборке.

### Обучение модели LogisticRegression

```python
%%time

# определяем модель
model_lr = LogisticRegression(
            random_state=RANDOM_STATE,
            class_weight='balanced'
        )

# формируем конвейер
pipeline_lr = make_pipeline(col_transformer, model_lr)

# определяем список и диапазон гиперпараметров для подбора
lr_grid_space = {
            'logisticregression__C': range (40,50,1)
        }

# подбираем гиперпараметры на кросс-валидации
lr_grid = GridSearchCV(
            pipeline_lr,
            param_grid=lr_grid_space,
            cv=5,
            scoring='roc_auc',
            n_jobs=-1
        )

# обучаем модель с лучшими гиперпараметрами
model_grid_lr = lr_grid.fit(features_train,target_train)

# получаем метрику модели с лучшими гиперпараметрами
final_score_lr = model_grid_lr.best_score_

# выводим результат
print('Гиперпараметры лучшей модели: '+str(model_grid_lr.best_params_))
print('Метрика ROC-AUC лучшей модели: '+str(final_score_lr))
```

### Обучение модели RandomForestClassifier

```python
%%time

# определяем модель
model_forest = RandomForestClassifier(random_state=RANDOM_STATE)

# формируем конвейер
pipeline_forest = make_pipeline(col_transformer, model_forest)

# определяем список и диапазон гиперпараметров для подбора
forest_grid_space = {
                'randomforestclassifier__n_estimators': [100, 500, 1000],
                'randomforestclassifier__max_depth': [5, 10, 15],     
                "randomforestclassifier__max_features": [*np.arange(0.1, 1.1, 0.2)]
              }

# подбираем гиперпараметры на кросс-валидации
forest_grid = GridSearchCV(
                pipeline_forest,
                param_grid=forest_grid_space,
                cv=5,
                scoring='roc_auc',
                n_jobs=-1
            )

# обучаем модель с лучшими гиперпараметрами
model_grid_forest = forest_grid.fit(features_train,target_train)

# получаем метрику модели с лучшими гиперпараметрами
final_score_forest = model_grid_forest.best_score_

# выводим результат
print('Гиперпараметры лучшей модели: '+str(model_grid_forest.best_params_))
print('Метрика ROC-AUC лучшей модели: '+str(final_score_forest))
```

### Обучение модели LGBMClassifier

```python
%%time

# определяем модель
model_lgbm = lgbm.LGBMClassifier(
                random_state=RANDOM_STATE, 
                verbose=-1
            )

# формируем конвейер
pipeline_lgbm = make_pipeline(col_transformer, model_lgbm)

# определяем список и диапазон гиперпараметров для подбора
lgbm_grid_space = {
                'lgbmclassifier__learning_rate': [0.1, 0.15, 0.2],
                'lgbmclassifier__max_depth': [8, 9, 10],
                'lgbmclassifier__num_leaves': [10, 11, 12],
                'lgbmclassifier__n_estimators': [350, 400, 450],
            }
 
# подбираем гиперпараметры на кросс-валидации
lgbm_grid = GridSearchCV(
                pipeline_lgbm,
                param_grid=lgbm_grid_space,
                cv=5,
                scoring='roc_auc',
                n_jobs=-1
            )

# обучаем модель с лучшими гиперпараметрами
model_grid_lgbm = lgbm_grid.fit(features_train,target_train)

# получаем метрику модели с лучшими гиперпараметрами
final_score_lgbm = model_grid_lgbm.best_score_

# выводим результат
print('Гиперпараметры лучшей модели: '+str(model_grid_lgbm.best_params_))
print('Метрика ROC-AUC лучшей модели: '+str(final_score_lgbm))
```

### Обучение модели CatBoostClassifier

```python
%%time

# определяем модель
model_catboost = CatBoostClassifier(
                    random_state=RANDOM_STATE,
                    verbose=False
                )

# определяем список и диапазон гиперпараметров для подбора
catboost_grid_space = {
                    'catboostclassifier__depth' : [1,2,3],
                    'catboostclassifier__learning_rate' : [0.135,0.14,0.145],
                    'catboostclassifier__iterations' : [1250, 1260, 1270]
                }

# формируем конвейер
pipeline_catboost = make_pipeline(col_transformer, model_catboost)

# подбираем гиперпараметры на кросс-валидации
catboost_grid = GridSearchCV(
                    pipeline_catboost,
                    param_grid=catboost_grid_space,
                    cv=5,
                    scoring='roc_auc',
                    n_jobs=-1,
                )

# обучаем модель с лучшими гиперпараметрами
model_grid_catboost = catboost_grid.fit(features_train,target_train)

# получаем метрику модели с лучшими гиперпараметрами
final_score_catboost = model_grid_catboost.best_score_

# выводим результат
print('Гиперпараметры лучшей модели: '+str(model_grid_catboost.best_params_))
print('Метрика ROC-AUC лучшей модели: '+str(final_score_catboost))
```
