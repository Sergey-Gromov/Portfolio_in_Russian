Значение метрики *RMSE* на тестовой выборке должно быть не больше 48.

6. Проверить данные на тестовой выборке и сделать выводы.



## Подготовка данных

### Загрузка библиотек и данных

Загрузка библиотек

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.stattools import adfuller

from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit

from catboost import CatBoostRegressor
from lightgbm import LGBMRegressor
```

Определение глобальных констант

```python
RANDOM_STATE = 12345
```

Загрузка данных

```python
pth1 = 'taxi.csv'
pth2 = '/datasets/taxi.csv'

if os.path.exists(pth1):
    data = pd.read_csv(pth1,index_col=[0], parse_dates=[0])
elif os.path.exists(pth2):
    data = pd.read_csv(pth2, index_col=[0], parse_dates=[0])
else:
    print('Что-то пошло не так')

#вывод параметров датасета для проверки    
display(data.shape)
display(data.info())
display(data.head())
display(data.describe())
```

### Исследование ряда на стационарность

Проведем оценку стационарности временного ряда количества заказов при помощи теста Дики — Фуллера.
При этом нулевая гипотеза формулируется как гипотеза о нестацинарности ряда, а альтернативная - как гипотеза о стационарности ряда:   
H_0: Временной ряд количества заказов нестационарен  
H_1: Временной ряд количества заказов стационарен  

Проверять будем для двух значений статистической значимости alpha: 1% и 5%

Определим функцию теста Дики — Фуллера

```python
def ad_fuller(test_data, alpha):
    
    st_test = adfuller(test_data)
    print('Тест для alpha =', alpha)
    
#   если значение с индексом 1 (p-value) меньше alpha, то ряд стационарный
    if st_test[1] < alpha:
        print('Результат: p-value =',st_test[1])    
        print('Нулевая гипотеза отвергнута – Временной ряд стационарен')
    else:
        print('Результат: p-value =',st_test[1])
        print('Нулевая гипотеза не отвергнута – Временной ряд нестационарен')
```

Проведем тест для двух значений статистической значимости alpha: 1% и 5%

```python
ad_fuller(data['num_orders'], 0.01)
ad_fuller(data['num_orders'], 0.05)
```

## Обучение моделей

### Создание признаков

```python
def make_features(data, max_lag, rolling_mean_size):
    df_new = data.copy()
    df_new['day'] = df_new.index.day
    df_new['dayofweek'] = df_new.index.dayofweek
    
    for lag in range(1, max_lag + 1):
        df_new['lag_{}'.format(lag)] = df_new['num_orders'].shift(lag)

    df_new['rolling_mean'] = df_new['num_orders'].shift().rolling(rolling_mean_size).mean()
    
    return df_new
```

```python
data_new = make_features(data, 25, 25)

# контроль результата
display(data_new.shape)
display(data_new.info())
display(data_new.head())
display(data_new.describe())
```

### Разделение данных на обучающую и тестовую выборки 

```python
# разделение выборок на обучающую и тестовую
train, test = train_test_split(data_new, shuffle=False, test_size=0.1)

# удаление строк с пропусками в обучающей выборке
train = train.dropna()

# выделение признаков и целевого признака
features_train = train.drop('num_orders', axis=1)
target_train = train['num_orders']
features_test = test.drop('num_orders', axis=1)
target_test = test['num_orders']

print(train.shape)
print(test.shape)
```

### Обучение моделей c применением кросс-валидации

Будем оценивать качество обучения на тренировочной выборке с для следующих моделей:
- модели Линейной регрессии
- модели градиентного бустинга LightGBM

#### Обучение модели Линейной регресссии

```python
lr_model = LinearRegression()
rmse_LR = (cross_val_score(lr_model,
                           features_train,
                           target_train,
                           cv=TimeSeriesSplit(n_splits=5),
                           scoring='neg_mean_squared_error').mean() * -1) ** 0.5
print('Средняя метрика  RMSE качества модели линейной регресии', rmse_LR)
```

#### Обучение модели градиентного бустинга LightGBM

```python
%%time

model_lgbm = LGBMRegressor(random_state=RANDOM_STATE) 

# определение списка и диапазона гиперпараметров для подбора
grid_space = [{'num_leaves':range (7, 13, 1),
               'n_estimators': range (130,150,1)}]

# подбор параметров с кросс-валидацией
lgbm_boost_grid = GridSearchCV(model_lgbm, 
                               param_grid=grid_space, 
                               cv=TimeSeriesSplit(n_splits=5), 
                               scoring='neg_mean_squared_error', 
                               n_jobs=-1)

model_grid_lgbm = lgbm_boost_grid.fit(features_train,target_train)

final_score_lgbm = abs(model_grid_lgbm.best_score_)**0.5

print('Гиперпараметры лучшей модели: '+str(model_grid_lgbm.best_params_))
print('Метрика RMSE лучшей модели: '+str(final_score_lgbm))
```

**Вывод**  
Наилучший результат по метрике RMSE = 25.7 показала модель градиентного бустинга LightGBM со следующими гиперпараметрами:  
- n_estimators = 135
- num_leaves = 9 
    
Именно эта модель признана лучшей и будет выбрана для проверки на тестовой выборке.

## Тестирование лучшей модели

### Проверка лучшей модели градиентного бустинга LightGBM на тестовой выборке

```python
predictions_test = model_grid_lgbm.predict(features_test)
print("Метрика RMSE лучшей модели:", mean_squared_error(target_test, predictions_test, squared=False))

```

**Визуализация предсказаний модели на фоне целевого признака**

```python
df_predictions_test = pd.DataFrame(predictions_test, index=target_test.index)

plt.rcParams.update({'figure.figsize':(10,5), 'figure.dpi':100})

plt.plot(target_test, label='целевой признак', color='blue')
plt.plot(df_predictions_test, label='предсказание модели', color='orange')

plt.title('Графики целевого признака и предсказания модели')
plt.xlabel('Время')
plt.ylabel('Количество заказов')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.show()
```

**Вывод:**  
Метрика RMSE = 43.3 лучшей модели на тестовой выборке меньше целевого уровня 48, следовательно модель удвлетворяет требованиям заказчика.  
Модель ошибается чаще всего на резких выбросах целевого признака и в целом дает более сглаженную картину.

### Проверка лучшей модели на адекватность

Проверка методом заполнения медианой

```python
pred_median = np.ones(target_test.shape) * target_train.median()
print("Метрика RMSE заполнения медианой:", mean_squared_error(target_test, pred_median, squared=False))
```

Проверка методом заполнения предыдущим значением

```python
pred_previous = target_test.shift(fill_value=0)
pred_previous.iloc[0] = target_train.iloc[-1]
print("Метрика RMSE заполнения предыдущим значением:", mean_squared_error(target_test, pred_previous, squared=False))
```

**Вывод:**  
Метод заполнения медианой показал результат RMSE = 87.2  
Метод заполнения предыдущим значением показал результат RMSE = 58.9  
Оба показателя значительно выше, чем результат нашей лучшей модели RMSE = 43.3  
Следовательно выбранная нами модель более, чем адекватна.
